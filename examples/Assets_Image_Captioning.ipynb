```python
import aempy
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Assets

```python
assets = aempy.Assets()
```

```python
img = assets.get_asset("/content/dam/we-retail/en/people/womens/women_6.jpg")
assets.display(img)
```

```python
imgs = assets.get_assets("/content/dam/we-retail/en", limit=1000)
imgs
```

```python
sizes = list(range(len(imgs)))
for i in range(len(imgs)):
    img = assets.get_asset(imgs[i])
    try:
        sizes[i] = img['jcr:content']['metadata']['dam:size']
    except:
        pass
```

```python
sizesKB = [x/1000 for x in sizes]
```

```python
plt.figure(figsize=(16, 6))
ax = sns.distplot(sizesKB);
ax.set_title('Image Size Distribution')
ax.set_xlabel('Size in KB')
ax.set_ylabel('Nb of images')
```

```python
plt.figure(figsize=(16, 6))
ax = sns.boxplot(sizesKB);
ax.set_title('Image Size Quartiles')
ax.set_xlabel('Size in kB')
```
